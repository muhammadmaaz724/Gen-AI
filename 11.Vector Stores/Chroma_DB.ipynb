```python
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
!pip install langchain_chroma
```

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader("Muhammad_Maaz_Resume.pdf")
docs = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 0
)
```

```python
text = splitter.split_documents(docs)
```

```python
print(text)
```

```python
vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory='CV_chroma_db',
    collection_name='sample'
)
```

```python
vector_store.add_documents(text)
```

```python
vector_store.get(include = ["embeddings","documents","metadatas"])
```

```python
similar = vector_store.similarity_search(query="what is his CGPA ?",k=1)
```

```python
print(similar[0].page_content)
```

```python
vector_store.similarity_search_with_score(query="what is his CGPA ?",k=2)
```

```python
vector_store.similarity_search(
    query="",
    filter={'source':'Muhammad_Maaz_Resume.pdf'}
    )
```

```python
from langchain.schema import Document

update_doc1 = Document(
    page_content="Maaz is a Good Boy",
    metadata={'source': 'His Mind'}
)
```

```python
vector_store.update_document(document_id="22643dfa-67ad-4fbe-b632-e1811d83bd79",document=update_doc1)
```

```python
vector_store.get(include = ["embeddings","documents","metadatas"])
```
