```python
from langchain_openai import OpenAIEmbeddings
from  dotenv import load_dotenv
```

```python
load_dotenv()
model = OpenAIEmbeddings(model='text-embedding-3-large',dimensions=32)
```

```python
query = "where is lahore ?"

documents = [
    "I am from Pakistan.",
    "Islamabad is the capital of Pakistan",
    "Lahore is in Pakistan",
    "out software house is in lahore"
]
```

```python
res = model.embed_query(query)
print(res)
```

```python
res = model.embed_documents(documents)
print(res)
```
